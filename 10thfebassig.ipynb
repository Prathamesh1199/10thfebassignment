```python

with open('data_scientist.txt', 'w') as file:
    file.write('I want to become a Data Scientist')

 
with open('data_scientist.txt', 'r') as file:
    content = file.read()


print(content)
```

```python
"explanation"
 In this program, I have first open a file named 'data_scientist.txt' in
write mode, and then use the write() method to write the string
"I want to become a Data Scientist" to the file. then close the file 
using the with statement.

Next, I open the same file in read mode using another with statement,
and use the read() method to read the content of the file. 
I store the content in the variable content.
```

```python
4: 1."read()"

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)

    
This code opens a file called 'example.txt' in read mode, reads the entire
contents of the file using the read() method, stores the contents in the
content variable as a string, and prints it to the console


2."readline()"

with open('example.txt', 'r') as file:
    line = file.readline()
    print(line)

This code opens a file called 'example.txt' in read mode, reads the 'first line' of
the file using the readline() method, stores the line in the line
variable as a string, and prints it to the console.    

3."readlines()"


with open('example.txt', 'r') as file:
    lines = file.readlines()
    print(lines)

    
This code opens a file called 'example.txt' in read mode, reads 'all lines' 
of the file using the readlines() method, stores the lines in the lines
variable as a list of strings, and prints it to the console. Each line in 
the list will be a separate string.









```

```python
5: In Python, the with statement is used in combination with the 
open() function to handle files.
 
>The with statement is used to create a context for the code to execute.
>when the context is exited, the file is automatically closed.
>Using the with statement simplifies file handling code.
>The with statement eliminates the need to explicitly close the file using 
the close() method.
>The with statement reduces the risk of forgetting to close the file and 
potentially causing errors in the code.
>The with statement ensures that the file is closed even if an error occurs
while the file is being read or written to.

```

```python
6:In Python, the write() and writelines() functions are used to write data
to a file.

> "write()": Writes a single string of data to a file.
>"writelines()": Writes multiple strings of data to a file,
 with each string on a separate line.
    
Examples:
 
# Example using write()
with open('example.txt', 'w') as file:
    file.write('Hello, world!\n')
    file.write('This is a test.\n')
    file.write('I am writing to a file.\n')


In this example, 
the write() function is used to write three lines of data to a file called
'example.txt'. The string \n is used to add a new line after each line of data.



# Example using writelines()
lines = ['Hello, world!\n', 'This is a test.\n', 'I am writing to a file.\n']
with open('example.txt', 'w') as file:
    file.writelines(lines)
    
In this example,
the writelines() function is used to write the same three lines of data to a
file called 'example.txt'.

The writelines() function takes a list of strings as its argument, with each string
on a separate line. The same \n character is used to add a new line
after each line of data.




```
